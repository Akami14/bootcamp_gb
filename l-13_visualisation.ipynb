```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\Оксана\Downloads\archive1\Stock-History\ZAll_Combine_Stock_Histry.csv')
df.head()
```

```python
df.shape
```

```python
df.info
```

```python
df.isnull().sum()
```

```python
sns.pairplot(df, hue="Company");
plt.figure(figsize=(20,20))
```

```python
df['year'] = [el[6:] for el in df['Date']]
df['month'] = [el[3:5] for el in df['Date']]
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(data=df, x="year", y="Volume");
```

```python
df['valat'] = ((df['High'] - df['Low'])/df['Open'])*100 # модуль изменения цены
df['pressantage_changes'] = ((df['Close'] - df['Open'])/df['Open'])*100 # изменение за день можеь асти может падать
df
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(data=df, x="year", y="valat");
```

```python
plt.figure(figsize=(30,10)) # plt.rcParams['figure.figsize']=(20,10)
sns.relplot(x = "year", y = "pressantage_changes", hue= "Company", data=df); 
```

```python
plt.figure(figsize=(20,10))
plt.scatter(df[df.Company == 'AAPL'].month, df[df.Company == 'AAPL'].valat)
plt.title('Вoлатильность Appl');
```

```python
plt.figure(figsize=(20,10))
plt.scatter(df[df.Company == 'NVDA'].month, df[df.Company == 'NVDA'].valat);
plt.title('Вoлатильность NVDA');
```

```python
plt.figure(figsize=(20,10))
plt.scatter(df[df.Company == 'TSLA'].month, df[df.Company == 'TSLA'].valat);
plt.title('Вoлатильность TSLA');
```

```python
sns.histplot(df["year"], binwidth=0.5, color='green');
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(x="Company", y="pressantage_changes", data=df);
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(x="year", y = "valat", data = df);
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x = "Company", y= "valat", hue = "Company", data = df)
plt.title('Зависимость волатиьность от компании');
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x = "year", hue ="year", data = df);
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(data = df, x = df['year'], y = df.Close, hue='Company');
```

```python
plt.figure(figsize=(25, 15))
sns.lineplot(data = df, x = df['year'], y = df.valat, hue='Company')
plt.title('Волатиьность');
```
